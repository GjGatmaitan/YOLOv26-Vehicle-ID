# PART 3 — Generate Performance Table

```python
metrics = model3.val()
```

Extract metrics:

```python
print("mAP50:", metrics.box.map50)
print("Precision:", metrics.box.mp)
print("Recall:", metrics.box.mr)
```

# Calculate F1 Score

```python
precision = metrics.box.mp
recall = metrics.box.mr

f1 = 2 * (precision * recall) / (precision + recall)
print("F1 Score:", f1)
```

# Summary Table

```python
import pandas as pd

data = {
    "Metric": ["mAP50", "Precision", "Recall", "F1 Score"],
    "Value": [
        metrics.box.map50,
        precision,
        recall,
        f1
    ]
}

df = pd.DataFrame(data)
df
```

```python
metrics1 = model1.val()
metrics2 = model2.val()
metrics3 = model3.val()
```

```python
def extract_metrics(metrics):
    precision = metrics.box.mp
    recall = metrics.box.mr
    map50 = metrics.box.map50
    f1 = 2 * (precision * recall) / (precision + recall)
    return precision, recall, map50, f1

p1, r1, m1, f1_1 = extract_metrics(metrics1)
p2, r2, m2, f1_2 = extract_metrics(metrics2)
p3, r3, m3, f1_3 = extract_metrics(metrics3)

metrics_table = pd.DataFrame({
    "Model": ["Model 1", "Model 2", "Model 3"],
    "mAP50": [m1, m2, m3],
    "Precision": [p1, p2, p3],
    "Recall": [r1, r2, r3],
    "F1 Score": [f1_1, f1_2, f1_3]
})

metrics_table
```
